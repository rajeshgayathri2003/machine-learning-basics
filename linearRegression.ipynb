```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("data.csv")

print(data)
```

```python
plt.scatter(data.data, data.value)
plt.show()
```

```python
def gradient_descent(m_now, b_now, points, L):
    m_gradient = 0
    b_gradient = 0
    
    n = len(points)
    
    for i in range(n):
        x = points.iloc[i].data
        y = points.iloc[i].value
        
        m_gradient += -(2/n)*x*(y - (m_now*x+b_now))
        b_gradient += -(2/n)*(y - (m_now*x+b_now))
        
    m = m_now - m_gradient * L
    b = b_now - b_gradient * L
    
    return m,b
        
```

```python
m = 0
b = 0
L = 0.001

epochs = 1000

for i in range(epochs):
    if i%50 == 0:
        print("{} Epochs".format(i))
    
    m,b = gradient_descent(m, b, data, L)
    
print(m,b)
        
plt.scatter(data.data, data.value, color = "black")
plt.plot(list(range(-6,6)), [m * x+ b for x in range(-6,6)], color="red")
plt.show()
    
```
